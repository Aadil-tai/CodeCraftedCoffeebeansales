# <font color='#4B197D'>Extracting Customer Information</font>

In the process of building this dashboard, the following formula was employed to extract customer information from the "Customers" sheet:

```excel
=+XLOOKUP(C2,customers!$A$1:$A$1001,customers!$B$1:$B$1001,,0)


**Formula Description:**

This formula enables us to efficiently retrieve customer names based on their unique Customer IDs. This enrichment of our dataset with valuable customer information is pivotal for in-depth analysis and reporting.


### Extracting Product Information Formula
```excel
=INDEX(products!$A$1:$G$49, MATCH(orders!$D3, products!$A$1:$A$49, 0), MATCH(orders!I$1, products!$A$1:$G$1, 0))


**Formula Description:**

This formula is utilized to enrich our sales data with comprehensive product information. Below is a breakdown of its components:

- **INDEX Function**: The INDEX function is employed to retrieve data from the "Products" sheet.

- **First MATCH Function**: This part of the formula matches the Product ID in our sales data with the Product IDs in the "Products" sheet to locate the corresponding row.

- **Second MATCH Function**: The second MATCH function is responsible for matching the header label (e.g., "Coffee Type" or "Roast Type") in our sales data with the headers in the "Products" sheet, thereby identifying the appropriate column.


## <font color="#4B197D">Creating Slicers and Timelines with Pivot Tables</font>

In the process of building this dashboard, I leveraged pivot tables to provide users with powerful tools for data exploration and analysis. Specifically, I created slicers and timelines to enhance the interactivity and usability of the dashboard.

### <font color="#4B197D">Pivot Tables for Data Summarization</font>

Before diving into slicers and timelines, let's highlight the role of pivot tables. I used pivot tables to summarize and aggregate the sales data, allowing me to present meaningful insights and trends. The pivot tables serve as the foundation for the dynamic charts and visualizations in the dashboard.

### <font color="#4B197D">Slicers for Data Filtering</font>

Slicers are interactive controls that empower users to filter and focus on specific aspects of the data. In this dashboard, I implemented three slicers:

1. **<font color="#4B197D">Roast Type Slicer:</font>** Users can select their preferred roast type (Dark, Light, or Medium) to narrow down the data and analyze sales trends for specific roasts.

2. **<font color="#4B197D">Size Slicer:</font>** The size slicer offers choices for bag sizes (0.2 kg, 0.5 kg, 1.0 kg, or 2.5 kg), allowing users to investigate how different bag sizes impact sales.

3. **<font color="#4B197D">Loyalty Card Slicer:</font>** With this slicer, users can filter customers based on their loyalty card status, helping to analyze the influence of loyalty programs on sales performance.

### <font color="#4B197D">Timelines for Date Selection</font>

Timelines provide users with the flexibility to choose specific date ranges for their analysis. In this dashboard, I incorporated a timeline control spanning from January 2019 to June 2022. Users can drag the handles to select their desired start and end dates, dynamically updating the data displayed in the dashboard.

### <font color="#4B197D">Interactive Charts and Visualizations</font>

The pivot tables, slicers, and timelines work seamlessly together to enable interactive exploration of the sales data. Users can observe trends over time, evaluate the impact of different factors (roast type, bag size, loyalty cards), and make data-driven decisions.

### <font color="#4B197D">Resetting Filters and Data Refresh</font>

To ensure a smooth user experience, I included options to clear slicer filters and refresh the data. The "Clear Filters" button allows users to reset slicers to their default state, while the "Refresh Data" button ensures that the dashboard always reflects the most up-to-date information.

By combining pivot tables, slicers, and timelines, I've created a robust and user-friendly platform for analyzing coffee sales data, helping users uncover insights and make informed decisions.

Please proceed to the sections below to begin exploring and interacting with the dashboard's features.


image.png

# <font color='#4B197D'>Dashboard Design</font>

The design of the **<font color='#4B197D'>Excel Dashboard Project</font>** was thoughtfully crafted to offer users an intuitive and visually pleasing experience while facilitating data analysis. Several key design elements were considered:

## <font color='#4B197D'>Layout and Organization</font>

- The dashboard features a clear and organized layout, ensuring easy navigation and information retrieval.
- Key components, including the timeline control, slicers, and bar charts, are logically placed for user convenience.

## <font color='#4B197D'>Color Scheme</font>

- A consistent color scheme was carefully chosen to enhance readability and create a unified visual identity for the dashboard.
- Colors were selected for their visual appeal without causing distraction.

## <font color='#4B197D'>Font and Typography</font>

- A legible font and appropriate typography were selected to ensure ease of reading for text and labels.
- Font sizes and styles were adjusted for clarity and hierarchy.

## <font color='#4B197D'>Data Visualization</font>

- Visualizations, such as the line chart and bar charts, were thoughtfully designed to effectively convey information.
- The line chart illustrates sales trends over time, while the bar charts highlight top customers and other pertinent data.

## <font color='#4B197D'>Interactivity</font>

- The dashboard incorporates interactive features, including the timeline control and slicers, enabling users to dynamically filter and explore data.
- Hover-over tooltips provide additional information and context for data points.

## <font color='#4B197D'>Consistency</font>

- Consistent design elements, such as buttons, icons, and labels, are employed throughout the dashboard, creating a unified user experience.

## <font color='#4B197D'>User-Friendly Interface</font>

- The dashboard was designed with user-friendliness in mind, ensuring individuals with varying levels of Excel expertise can effectively navigate and utilize its features.

## <font color='#4B197D'>Responsiveness</font>

- The dashboard is designed to adapt to different screen sizes and resolutions, maintaining user-friendliness on various devices.

The overall dashboard design strives to strike a balance between aesthetics and functionality, providing users with a visually appealing and user-friendly platform for exploring and analyzing coffee sales data.


# <font color='#4B197D'>Interactivity</font>

One of the primary strengths of the **<font color='#4B197D'>Excel Dashboard Project</font>** lies in its interactive features, empowering users to actively engage with and explore the data. The following interactive elements have been seamlessly integrated into the dashboard:

## <font color='#4B197D'>Timeline Control</font>

- The timeline control enables users to select a specific date range within the dataset, with an intuitive interface for choosing start and end dates.
- By selecting different time periods, users can observe how sales data evolves over time, identifying trends and patterns.

## <font color='#4B197D'>Slicers for Data Filtering</font>

- The dashboard incorporates three slicers designed to facilitate data filtering and segmentation:
  - **Roast Type Slicer:** Users can filter data by selecting from Dark, Light, or Medium roast types.
  - **Size Slicer:** Users can filter data based on bag size, choosing from 0.2 kg, 0.5 kg, 1.0 kg, and 2.5 kg.
  - **Loyalty Card Slicer:** Users can filter data to include or exclude customers with loyalty cards.
- These slicers empower users to customize their analysis and view the most relevant data.

## <font color='#4B197D'>Bar Charts Interaction</font>

- The bar charts dynamically respond to user selections made through the timeline and slicers. Real-time updates in the bar charts reflect the filtered data.
- This real-time interaction enables users to explore how different factors, such as roast type or bag size, impact sales performance.

## <font color='#4B197D'>Tooltip Information</font>

- Hover-over tooltips provide valuable additional information and context for data points on the bar charts. Users can gain insights into specific sales values and customer details by simply hovering over data bars.

## <font color='#4B197D'>User Guidance</font>

- The dashboard includes clear instructions and guidance to help users understand how to effectively use the interactive features.
- Users are encouraged to explore different scenarios by adjusting the timeline and slicers to gain deeper insights into the data.

The interactive elements integrated into the dashboard enhance the user experience by providing flexibility in analyzing data from various perspectives. Users can interact with the dashboard to answer specific questions and make well-informed, data-driven decisions effectively.


image.png

## <font color="#007BFF">Dynamic Data Updates</font>

<font color="#6C757D">As demonstrated in the dashboard</font>, you can observe how the selection of the time frame from February to June 2022 dynamically updates key data points. These updates include:

- <font color="#28A745">Total Sales:</font> See how the total sales figures change in response to your selected time frame.

- <font color="#FFC107">Sales by Country:</font> Explore how sales distribution across different countries evolves in real-time.

- <font color="#DC3545">Top 5 Customers:</font> Observe the list of the top-performing customers, which dynamically adjusts based on your criteria.

This dynamic data updating feature empowers users to explore and analyze the data in real-time, providing a responsive and interactive experience tailored to their preferences.


# <font color='#4B197D'>Instructions for Use</font>

To effectively utilize the **<font color='#4B197D'>Google Colab Notebook</font>** and gain insights into coffee sales data, follow these step-by-step instructions:

1. **Selecting a Date Range with the Timeline Control:**

   - Use the timeline control located  to choose a specific date range within the available timeline (January 2019 to June 2022).
   - Click and drag the timeline handles to adjust the start and end dates to your desired range.
   - Observe how the data in the dashboard updates to reflect the selected timeline.

2. **Filtering Data Using Slicers:**

   - Utilize the slicers on the dashboard to filter data based on your preferences:
     - **Roast Type Slicer:** Click on the roast type (Dark, Light, or Medium) to filter data by roast type.
     - **Size Slicer:** Select bag sizes (0.2 kg, 0.5 kg, 1.0 kg, or 2.5 kg) to filter data by bag size.
     - **Loyalty Card Slicer:** Choose to include or exclude customers with loyalty cards.
   - The charts and visualizations will dynamically update to display the filtered data.

3. **Interacting with Bar Charts:**

   - Explore the bar charts to gain insights:
     - **Top 5 Customers Bar Chart:** Hover over individual bars to view sales values for the top 5 customers during the selected timeline and filtered slicers.
     - **Country Bar Chart:** This chart displays county-wise sales data. Hover over bars to see specific sales figures for each county.

4. **Data Exploration:**

   - Use the interactive features to explore sales data and answer specific questions about sales trends, customer behavior, and more.
   - Experiment with different combinations of timeline selections and slicer filters to analyze data from various angles.

5. **Resetting Filters:**

   - To clear slicer filters and revert to the default view, click the "Clear Filters" button.

6. **Data Refresh:**

   - If your data source is regularly updated, ensure that you periodically refresh the dashboard to reflect the most current data. Click the "Refresh Data" button.

7. **Assistance and Support:**

   - If you encounter any issues or have questions while using the dashboard, please don't hesitate to reach out for assistance. Contact aadiltaitech@gmail.com.

These instructions are designed to help you effectively use the dashboard's interactive features and gain meaningful insights from the sales data. Feel free to explore and customize the dashboard to meet your specific analysis and reporting needs.


[Dashboard video link](https://drive.google.com/file/d/1tYgT9vdiQpKXeuhBmxzu4HxCWC3KimUH/view?usp=drive_link)
